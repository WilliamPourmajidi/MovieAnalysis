# a

Write a simple implementation of a least-squares solution to linear regression that applies an iterative update to adjust the weights. Demonstrate the success of your approach on the sample data loaded below, and visualize the best fit plotted as a line (consider using linspace) against a scatter plot of the x and y test values.


```python
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
from matplotlib import pyplot as plt 


# Load the diabetes dataset
diabetes = datasets.load_diabetes()


# Use only one feature
diabetes_X = diabetes.data[:, np.newaxis, 2]

# Split the data into training/testing sets
diabetes_X_train = diabetes_X[:-20]
diabetes_X_test = diabetes_X[-20:]


# Split the targets into training/testing sets
diabetes_y_train = diabetes.target[:-20]
diabetes_y_test = diabetes.target[-20:]


```

```python


#### Approach 1 - Using Ordinary Least Square Method - Iterating through to find values for coefficient and y-intercept in y = mx + b  #######

# Calculating the mean of X and y
diabetes_X_train_mean = np.mean(diabetes_X_train)
diabetes_y_train_mean = np.mean(diabetes_y_train)
# print(diabetes_X_train_mean)
# print(diabetes_y_train_mean)

len_of_data = len(diabetes_X_train)

# Using the formula to calculate b1 and b2   https://mubaris.com/posts/linear-regression/
numer = 0
denom = 0
# Iterating to find the wights (in this case, Y-intercept and coefficient 
for i in range(len_of_data):
    numer += (diabetes_X_train[i] - diabetes_X_train_mean) * (diabetes_y_train[i] - diabetes_y_train_mean)
    denom += (diabetes_X_train[i] - diabetes_X_train_mean) ** 2
b1 = numer / denom
b0 = diabetes_y_train_mean - (b1 * diabetes_X_train_mean)

# Print the details of the regression line
print(f'By Using Approach #1, for the line y = b0 + b1x , the coefficient (slope) is {b1}, and the y-intercept is {b0}')

# Plotting Values and Regression Line
min_x = np.min(diabetes_X_train)
max_x = np.max(diabetes_X_train)

# Calculating line values x and y
x = np.linspace(min_x, max_x, 10)
y = b0 + b1 * x

plt.figure(figsize=(12, 12))
plt.title("Linear Regression for Diabetes Data Using Ordinary Least Square Method", color='white')
plt.scatter(diabetes_X_test, diabetes_y_test, color='blue')
plt.plot(x, y, color='red', linewidth=2)
plt.show()

print('-------------------- Approach#2 (Etcetra) just for the sake of comparison -----------')
#### Approach 2 -  Using Scikit Learn Linear Regression function  (just to compare the result #######
# Instantiating a linear Regression object
lnr_regression = linear_model.LinearRegression()

# Fitting the training sets to the linear regression object (model)
lnr_regression.fit(diabetes_X_train, diabetes_y_train)

# defining a variable and assigning predicted values to it
diabetes_y_hat = lnr_regression.predict(diabetes_X_test)

# print('Details of the Regression \n  Coefficient (Slope) : \n', lnr_regression.coef_)
print(f'Details of the Regression (for approach #2) is presented below:\n  '
      f'Coefficient (Slope) : {lnr_regression.coef_}\n'
      f'Y-intercept : {lnr_regression.intercept_}\n'
      f'Mean squared error : {mean_squared_error(diabetes_y_test, diabetes_y_hat)}\n'
      f'Variance score: {r2_score(diabetes_y_test, diabetes_y_hat)}'
      )

# Plot outputs
plt.figure(figsize=(12, 12))
plt.title("Linear Regression for Diabetes Data Using Scikit Linear Regression", color="white")
plt.scatter(diabetes_X_test, diabetes_y_test, color='blue')
plt.plot(diabetes_X_test, diabetes_y_hat, color='red', linewidth=2)
plt.grid(True)
plt.show()



```

# a

Write a simple version of the basic algorithm for k-means clustering. Simple here means the core of the algorithm and not optimizations or extensions you might find in standard python libraries. Typically you might rely on a standard library for doing this, but it helps to see the core by manipulating the data and labels by hand as practice for numerical python and how to frame the algorithm.

```python
# Import packages
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()  # for plot styling
import numpy as np


# Generate Samples
from sklearn.datasets.samples_generator import make_blobs
X, y_true = make_blobs(n_samples=300, centers=4,
                       cluster_std=0.60, random_state=0)
plt.scatter(X[:, 0], X[:, 1], s=50);

###############################################
# YOUR CODE GOES HERE
# Put some code to find clusters here
# Assign the clusters and labels in your code

# def dist(a, b, ax=1):
#     return np.linalg.norm(a - b, axis=ax)

from sklearn.metrics import pairwise_distances_argmin
def manual_clustering(X, number_of_clusters, rseed=2):
    # Randomly choose clusters using np.random
    rnge = np.random.RandomState(rseed)
    a = rnge.permutation(X.shape[0])[:number_of_clusters]
    centers = X[a]

    while True:
        #  Assign labels based on the outcome of pairwise_distances_argmin which yeilds to the closest center
        labels = pairwise_distances_argmin(X, centers)
        #         labels = dist(X, centers)

        # Cacluate the mean of points and form new candidate centers from them
        candidate_centers = np.array([X[labels == a].mean(0)
                                      for a in range(number_of_clusters)])

        if np.all(centers == candidate_centers):
            break
        centers = candidate_centers

    return centers, labels

centers, labels = manual_clustering(X, 4)
plt.scatter(X[:, 0], X[:, 1], c=labels,
            s=50, cmap='viridis');


###############################################


# Uncomment to display clusters and cluster centers
plt.scatter(X[:, 0], X[:, 1], c=labels,
           s=50, cmap='viridis');
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5);
```

YOUR EXPLANATION HERE, provide an example for one user id below.
<font color='lightgreen'>
First, we will need to drop extra attributes that we no longer need for a recommender system. Among all the attributes in the meta_data, we only need  'id', 'original_title', 'original_language' and as for ratings_small data set, we need 'userId', 'movieId', 'rating'.  

As rating and userId comes from the ratings_small data set and information about movie comes meta_data data set, one needs to join or merge them. The idea is very similar to the inner joins that are common in relational databases (SQL).  To create a common id (similar to foreign key), we need to rename the id in meta_data to userID. 

For the sake of simplicity, we focus on movies that their original_language is English. One needs to consider that an attribute such as spoken_languages can be constructed by listing all the movies that the user has watched and improve the attribute quality by considering the country of the user (for future work!). 

Now, we are almost ready to merge our two panda dataframes, yet, after checking the dtype for attributes we notice that the movieId of ratings_small is int64 but for the meta_data is of type object. We change the type of meta_data userId to int64 as well.  Now, userId attribute exist on both data frame and is of the same type, hence, we proceed with the merge. 

After merge, we construct a matrix of all the movies as columns and all the users as rows and put the ranking  of the a user on a movie at the intersection of these two attributes. 

We chose Pearson Correlation, as  a measure of the linear correlation between two attributes( user and movie). Two functions are created and exposed. One to calculate Pearson Correlation between two attributes and another one to recommend movies based on the give movie name, matrix name, and number of expected recommendations.

For a given user, lets say userId, we build an array of the movies that he or she has watched (and therefore might ranked!).  we take the first movie of this array and we pass it onto the function that recommends a configurable number of recommended movies. 

While this process is of passing the movie to function is done manually in this work, it can be easily automated. 
Once the function returns the list of recommended movies, we will need to remove the watched movies from it and show the rest to the user.

```python

meta_data = pd.read_csv('movies_metadata.csv',
                        low_memory=False)  # You may wish to specify types, or process columns once read
ratings_small = pd.read_csv('ratings_small.csv')
# Dropping extra attributes that are not needed
ratings_small = ratings_small[['userId', 'movieId', 'rating']]
meta_data = meta_data[['id', 'original_title', 'original_language']]

# renaming the colum of id in meta_data data set to movieID
# so it can be merged with ratings_small data set on this attribute
meta_data = meta_data.rename(columns={'id': 'movieId'})

# For the sake of simiplicity, we only focus on English movies
meta_data = meta_data[meta_data['original_language'] == 'en']
# Displaying the current state of our data sets
print("Here is how meta_data data set header and information and shape")
meta_data.head()
meta_data.info()
meta_data.shape

print("Here is how ratins_small data set header and information and shape")
ratings_small.head()
ratings_small.info()
ratings_small.shape

# movieId of ratings is int64 but for the meta_data is of type object, need to make them the same
meta_data.movieId = pd.to_numeric(meta_data.movieId, errors='coerce')
ratings_small.movieId = pd.to_numeric(ratings_small.movieId, errors='coerce')

# checking to make sure they are now both int64
meta_data.info()
ratings_small.info()

# We now merge the twp panda data frame on unique movieId (similar to inner join in relational databases)
merged_dfs = pd.merge(ratings_small, meta_data, on='movieId', how='inner')
merged_dfs.head(100)


# This will be used in the recommender function
recommender_matrix = merged_dfs.pivot_table(index='userId', columns='original_title', values='rating')
print("The following matrix will be used for recommendation in the future")
recommender_matrix.head()

# Pearson Correlation as a measure of liner correlation
def pearsonR(s1, s2):
    s1_c = s1 - s1.mean()
    s2_c = s2 - s2.mean()
    return np.sum(s1_c * s2_c) / np.sqrt(np.sum(s1_c ** 2) * np.sum(s2_c ** 2))

# The following function takes a moviename, and the recommender_matrix , as well as the number of
# expected recommendations and returns the recommended movies based on calculated Pearson Correlation

def movie_recommender(movie, M, n):
    reviews = []
    for title in M.columns:
        if title == movie:
            continue
        cor = pearsonR(M[movie], M[title])
        if np.isnan(cor):
            continue
        else:
            reviews.append((title, cor))

    reviews.sort(key=lambda tup: tup[1], reverse=True)
    return reviews[:n]


user_001_watched_movies = ['Wild Wild West', '12 Angry Men', 'Donnie Darko', 'Wizards of the Lost Kingdom', 'Wild in the Country']
# The following imaginary userId can be easily replaced by a real user from the list, however, the list of watched movies has to be extracted
recommendation_for_user_001 = movie_recommender('Wild Wild West', recommender_matrix, 10)  # top 10 recommendations
print("Here is the recommended movies for user_001 (an imaginary userId)")
recommendation_for_user_001

# We will remove a movie that is recommended to the user but the user has seen it before
remove_watched_from_recommended_list = [r for r in recommendation_for_user_001 if r[0] not in user_001_watched_movies]
# This is the subtle difference that doesn't get on my nerves...
print("Here is the recommended movies for user_001 the he or she has never seen before!")
remove_watched_from_recommended_list


```
